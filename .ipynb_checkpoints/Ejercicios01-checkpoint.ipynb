### Desarrolle el Reto

```python
from ipyleaflet import Map, MagnifyingGlass, basemaps, basemap_to_tiles, Marker, AwesomeIcon
import pandas as pd
import ipywidgets as widgets
```

```python
datos = pd.read_json("data/ocw_universidades.json")
datos
```

```python
datosFinales = datos[["universityName", "universityLat", "universityLong"]]
datosFinales
```

```python
lista_columnas = ['nombre_U', 'latitud', 'longitud']
datosFinales.columns = lista_columnas
datosFinales
```

```python
datosDiccionario = datosFinales.to_dict("records")
datosDiccionario
```

```python
datos_tuplas = [(i['nombre_U'], (i['latitud'], i['longitud'], i['nombre_U'])) for i in datosDiccionario]
datos_tuplas
```

```python
datos_tuplas[0][0] # Posición cero de la tupla
```

```python
datos_tuplas[0][1] # Posición uno de la tupla
```

```python
universidades = widgets.Dropdown(
    options = datos_tuplas,
    value = (-1.294246, 36.786794, 'African Virtual University'),
    description='Universidad:',
)

universidades # Caja con nombre de las universidades
```

```python
latitud = universidades.value[0]
longitud = universidades.value[1]
nombre_u = universidades.value[2]
centro = (latitud, longitud)
if(not pd.isna(latitud) and not pd.isna(longitud)):
    mapa = Map(center=centro, zoom=5, basemap=basemaps.Stamen.Terrain)
    icono = AwesomeIcon(
        name='bus',
        marker_color='red',
        icon_color='white',
        spin=False
    )
    c = (latitud, longitud)
    marker = Marker(icon=icono, location=c, draggable=False, title=nombre_u, alt=nombre_u) 
    mapa.add_layer(marker)
else:
    mapa = Map(center=(0,0), zoom=2, basemap=basemaps.Stamen.Terrain)
    print("No disponemos de la localización de dicha universidad")
    
mapa
```
